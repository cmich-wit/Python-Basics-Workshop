Take this list and print it in the opposite direction (cherry, banana, apple)

```python
fruits = ["apple", "banana", "cherry"]
for int in range(len(fruits)-1, -1, -1):
    print(fruits[int])
```

Final challenge: create rock paper scissors game

```python
import random
list = ['rock', 'paper', 'scissors']
flag = 'Y'
while flag == 'Y':
  computer_choice = random.choice(list)
  user_choice = input("Enter rock, paper, or scissors:")
  print("Computer chose", computer_choice)
  if computer_choice == user_choice:
      print("It's a tie")
  if computer_choice == 'rock' and user_choice == 'paper':
      print("You win")
  if computer_choice == 'paper' and user_choice == 'scissors':
      print("You win")
  if computer_choice == 'scissors' and user_choice == 'rock':
      print("You win")
  if computer_choice == 'rock' and user_choice == 'scissors':
      print("You lose")
  if computer_choice == 'scissors' and user_choice == 'paper':
      print("You lose")
  if computer_choice == 'paper' and user_choice == 'rock':
      print("You lose")
  flag = input('Do you want to play again? Y or N: ')
  flag = flag.upper()
  print('-----')
```
