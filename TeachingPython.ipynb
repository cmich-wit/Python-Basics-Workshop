Lists, tuples, sets, dictionaries

```python
# Declaring and initializing a list
#List items are ordered, changeable, and allow duplicate values.
list = [1, 1, 2, 2, 5]

# Accessing list elements
first_element = list[0]
print(first_element) # first_element will be 1
```

```python
#delete item from list
list.remove(5)
print(list)
```

```python
#declaring and initializing a tuple
#Tuple items are ordered, unchangeable, and allow duplicate values.
tuple(1,1,2,2,5)

# Accessing tuple elements
first_element = tuple[0]
print(first_element) # first_element will be 1
```

```python
#what happens if you try to delete a item from unchangeable tuple
tuple.remove(5)
print(tuple)
```

```python
#declaring and initializing a set
#Set items are unordered and do not allow duplicate values.
set = {1, 1, 2, 2, 5}

print(set)
```

```python
#declaring and initializing a dictionary
#Dictionary items are ordered and do not allow duplicates keys.
# key : value
dictionary = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
}
print(dictionary)
```

Creating a class in python

```python
#creating a class in python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

#instantiating object
person1 = Person("Alice", 25)
person1.greet()
```

Compare and calculate

```python
#comparisons
x = 2
print(x == 2) # prints out True
print(x != 3) # prints out True
print(x < 3) # prints out True
print(x > 3) # prints out False
```

```python
#calculations
print (x + 1)   # Addition; prints "3"
print (x - 1)   # Subtraction; prints "1"
print (x * 2)   # Multiplication; prints "2"
print (x ** 2)  # Exponentiation; prints "4"
print (x % 2)  # Modulus; prints "0"
```
